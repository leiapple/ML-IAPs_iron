# Python code for generating GAP grid plots

```python
# general imports
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

## RMSE Training Energy

```python
# all data related to this plot are given in folder `Fig2_data`
# load the RMSE data
my_data = np.genfromtxt('./gap_data/Ermse_train.csv', delimiter=',')
data_rev = np.zeros([12,12])
for i in range(12):
    for j in range(12):
        data_rev[i,j] = my_data[11-i,j] * 1000
# load the computational speed data
speed = np.genfromtxt('./gap_data/gap_timing.txt')

f1 = plt.figure(figsize=(10,10))
ax = plt.gca()
ec = plt.imshow(data_rev[:,:],interpolation='none',cmap='viridis', vmin=0, vmax=20)

for i in range(12):
    for j in range(12):
        ax.text(j-0.2,i+0.2,int(speed[i,j]),color='red',fontsize=10)

ax.set_xticks(np.arange(0,12,1))
ax.set_xticklabels(np.arange(1,13,1))
ax.set_yticks(np.arange(0,12,1))
ax.set_yticklabels(np.arange(12,0,-1))
ax.set_ylabel(r'$l_{\mathrm{max}}$')
ax.set_xlabel(r'$n_{\mathrm{max}}$')
              
c = plt.colorbar(ec, shrink=0.5)
c.set_ticks(np.linspace(0,20,5))
c.set_label('Training energy error (meV/atom)', rotation=270, labelpad=24.0)

plt.show()
```

## RMSE Testing Energy

```python
# all data related to this plot are given in folder `Fig2_data`
# load the RMSE data
my_data = np.genfromtxt('./gap_data/Ermse_test.csv', delimiter=',')
data_rev = np.zeros([12,12])
for i in range(12):
    for j in range(12):
        data_rev[i,j] = my_data[11-i,j] * 1000

f1 = plt.figure(figsize=(10,10))
ax = plt.gca()
ec = plt.imshow(data_rev[:,:],interpolation='none',cmap='viridis', vmin=0, vmax=20)

for i in range(12):
    for j in range(12):
        ax.text(j-0.2,i+0.2,int(speed[i,j]),color='red',fontsize=10)

ax.set_xticks(np.arange(0,12,1))
ax.set_xticklabels(np.arange(1,13,1))
ax.set_yticks(np.arange(0,12,1))
ax.set_yticklabels(np.arange(12,0,-1))
ax.set_ylabel(r'$l_{\mathrm{max}}$')
ax.set_xlabel(r'$n_{\mathrm{max}}$')
              
c = plt.colorbar(ec, shrink=0.5)
c.set_ticks(np.linspace(0,20,5))
c.set_label('Test energy error (meV/atom)', rotation=270, labelpad=24.0)

plt.show()
```

## RMSE Training Force

```python
# all data related to this plot are given in folder `Fig2_data`
# load the RMSE data
my_data = np.genfromtxt('./gap_data/Frmse_train.csv', delimiter=',')
data_rev = np.zeros([12,12])
for i in range(12):
    for j in range(12):
        data_rev[i,j] = my_data[11-i,j] 

f1 = plt.figure(figsize=(10,10))
ax = plt.gca()
ec = plt.imshow(data_rev[:,:],interpolation='none',cmap='viridis', vmin=0.04, vmax=0.2)

for i in range(12):
    for j in range(12):
        ax.text(j-0.2,i+0.2,int(speed[i,j]),color='red',fontsize=10)

ax.set_xticks(np.arange(0,12,1))
ax.set_xticklabels(np.arange(1,13,1))
ax.set_yticks(np.arange(0,12,1))
ax.set_yticklabels(np.arange(12,0,-1))
ax.set_ylabel(r'$l_{\mathrm{max}}$')
ax.set_xlabel(r'$n_{\mathrm{max}}$')
              
c = plt.colorbar(ec, shrink=0.5)
c.set_ticks(np.linspace(0.04,0.2,5))
c.set_label('Training force error (eV/Å)', rotation=270, labelpad=24.0)

plt.show()
```
