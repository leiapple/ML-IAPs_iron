# Python code for plotting Figure 2

The code for generating the convergence plot for all ML-IAPs.

```python
# general import
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

font = {'family' : 'helvetica',
'weight' : 'normal',
'size' : 20}
matplotlib.rc('font', **font)

# define a set of colors
cc1='#F7AC08'
cc2='#039069'
cc3='#0072B2'
cc4='#D55E00'
cc5='#CC79A7'
```

## GAP plot

```python
# training and testing RMSE on energy and forces 
# Data are for M=400, 800, 1600, 3200,6400

gap_Ermse_train = np.array([0.00562611, 0.00398443, 0.00289272, 0.00252179, 0.00245326])
gap_Ermse_test = np.array([0.00553768, 0.0052363 , 0.0032033 , 0.00284293, 0.00276007])
gap_Frmse_train = np.array([0.12354761, 0.10161644, 0.07588917, 0.06920102, 0.06561962])
gap_Frmse_test = np.array([0.11345776, 0.09074979, 0.06698211, 0.06488045, 0.05657742])


fig, (ax) = plt.subplots(nrows=1, ncols=1, figsize=(10,10))

m = np.array([400,800,1600,3200,6400])

ax.plot(m,gap_Ermse_train*1000,lw=3,color=cc2,marker='o',markersize=12,label='Training')
ax.plot(m,gap_Ermse_test*1000,c=cc2,lw=3,ls='--',marker='o',markersize=12,fillstyle='none',label='Testing')

ax.grid(axis='x', which='both', ls=':')
ax.grid(axis='y', which='major', ls=':')
ax.set_ylabel('Energy RMSE, (meV/atom)')
ax.set_xlabel('No. of representative LAE')
ax.legend()

plt.show()
```
